# Prestige Ranking

```python
df['Training on Job'].unique()
```

```python
df.loc[df['Entry level education'] == 'No formal educational credential','Education weight'] = 1
df.loc[df['Entry level education'] == 'High school diploma or equivalent','Education weight'] = 2
df.loc[df['Entry level education'] == "Bachelor's degree",'Education weight'] = 3
df.loc[df['Entry level education'] == "Associate's degree",'Education weight'] = 3.5
df.loc[df['Entry level education'] == "Master's degree",'Education weight'] = 4.5
df.loc[df['Entry level education'] == 'Postsecondary nondegree award','Education weight'] = 3.5
df.loc[df['Entry level education'] == 'Some college, no degree','Education weight'] = 2.5
df.loc[df['Entry level education'] == 'Doctoral or professional degree','Education weight'] = 6
```

```python
df.loc[df['Work Experience'] == 'None','Experience weight'] = 1
df.loc[df['Work Experience'] == '5 years or more','Experience weight'] = 3
df.loc[df['Work Experience'] == "Less than 5 years",'Experience weight'] = 2
```

```python
df.loc[df['Training on Job'] == 'None','Training weight'] = 1
df.loc[df['Training on Job'] == 'Apprenticeship','Training weight'] = 2
df.loc[df['Training on Job'] == 'Internship/residency','Training weight'] = 3
df.loc[df['Training on Job'] == 'Short-term on-the-job training','Training weight'] = 3.5
df.loc[df['Training on Job'] == 'Moderate-term on-the-job training','Training weight'] = 4.5
df.loc[df['Training on Job'] == 'Long-term on-the-job training','Training weight'] = 6.5
```

```python
df.head()
```

```python
df['Prestige']=(df['Education weight']*df['Training weight']*df['Experience weight']*df['Mean Annual Salary'])
df.head()
```

```python
df['Education weight']=df['Education weight'].astype(float)
df['Experience weight']=df['Experience weight'].astype(float)
df['Training weight']=df['Training weight'].astype(float)
df['Prestige']=df['Prestige'].astype(float)
```

```python
pres = df.sort_values(by='Prestige', ascending=False)
pres.reset_index(inplace=True, drop=True)
```

```python
pres.to_csv('prestige.csv')
```

```python
df['popularity'] = (df['TOT_EMP']*df['Prestige']*df['Employement change in 10 years']*(df['16 to 19 years']+df['20 to 24 years']+df['25 to 34 years']))/(df['Training weight']*df['Education weight']*df['Experience weight'])
df.head()
```

```python
p = df.sort_values(by='popularity', ascending=False)
#p = p[p['O_GROUP']=='detailed']
p.reset_index(inplace=True, drop=True)
```

```python
p.to_csv('general_job_popularity.csv')
```

```python
final = p.sort_values(by='OCC_CODE')
final.reset_index(inplace=True, drop=True)
final.to_csv('CSE519_Job_Title_dataset.csv')
```
